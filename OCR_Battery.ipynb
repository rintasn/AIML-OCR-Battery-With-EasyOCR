```python
def read_and_show(image_name,color):
    image = cv2.imread(image_name)
    result = reader.readtext(image)
    for (bbox,text,prob) in result:

#         print(bbox,text,prob)
        (top_left, top_right, bottom_right, bottom_left) = bbox
        top_left = tuple([int(val) for val in top_left])
        bottom_right = tuple([int(val) for val in bottom_right])
        image =  cv2.rectangle(image, top_left, bottom_right, color, 2)
        image = cv2.putText(image,text,(top_left[0], top_left[1] - 10),0,2,color,2)
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
read_and_show("/content/drive/MyDrive/OCR_dibimbing/battery1.png",(0,255,0))
```
